```python
#plot a line plot ('JUMPSTART Attendees')

fig = plt.figure()

ax = fig.add_subplot()
ax.plot(df_ju.Year, df_ju.Attendees)
ax.set_title("JUMPSTART Attendees")
ax.set_xlabel("Year")
ax.set_ylabel("Attendees")
```

```python
#pie chart (What year from 2020-2022 was students' first year on team), where the slices will be ordered and plotted counter-clockwise

labels = ['Before COVID', '2021', '2022']
sizes = [8, 9, 15]

#explode out pie piece by offsetting it a greater amount

explode = (0.125, .03, 0.03)

fig, ax = plt.subplots()
ax.pie(sizes,
       #hexadecimal values
       colors=['#011f4b','#03396c', '#005b96'],
       explode=explode,
       labels=labels,
       autopct='%1.1f%%',
              #controls # of decimal places each % has
       shadow=False,
              #if in light mode shadow adds shadow under pie chart
       startangle=90)
              #degree in which pie chart starts, changing value spins chart
ax.axis('equal')  # equal aspect ratio ensures the pie chart is circular.
ax.set_title('First Year of FRC for 2022 Team Members')


plt.show()
```

```python
#no explode
#pie chart (What year from <2020-2022 was students first year, no explosion), where the slices will be ordered and plotted counter-clockwise

labels = ['Before COVID', '2021', '2022']
sizes = [8, 9, 15]
       #data values manually put in

fig, ax = plt.subplots()
ax.pie(sizes,
       colors=['#011f4b','#03396c', '#005b96'],
              #hexadecimal values
       labels=labels,
       autopct='%1.1f%%',
       startangle=90)
ax.axis('equal')  #equal aspect ratio ensures the pie chart is circular.
ax.set_title('First Year of FRC for 2022 Team Members')


plt.show()
```

```python
#side-by-side bar chart for mentor to student ratio
#manually enter data points into array

A = np.array([30,24,32,45,44,40,36,35,28,32])
B = np.array([12,10,12,10,13,12,12,16,8,18])
X = np.arange(10)
 
plt.barh(X, A, color = 'r')
    #- on 'A' flips graph if B is positive
plt.barh(X, -B, color = 'b')
plt.title("Mentor Student Ratio")
plt.ylabel("Year")
plt.xlabel("Mentors vs Students")


plt.show()
```

```python
#stacked bar chart for mentor to student ratio
#manually enter data points into array

A = np.array([30,24,32,45,44,40,36,35,28,32])
B = np.array([12,10,12,10,13,12,12,16,8,18])
X = np.arange(10)
 

plt.barh(X, A, color = 'r', edgecolor='white')
    #edgecolor is the outline
plt.barh(X, B, color = 'b')
    #when 'A' and 'B' are both - graph is flipped
plt.title("Mentor Student Ratio")
plt.ylabel("Year")
plt.xlabel("Member Number")


plt.show()
```
